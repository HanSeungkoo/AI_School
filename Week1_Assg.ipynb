##2. 시리
* 시리도 꼬박꼬박 사용하게되는 편리한 서비스이다. 나의 경우엔 이 세 가지 기능 사용이 습관화 되었다. 
  1. 날씨를 물어볼 때 - 오늘 비와? , 일주일 날씨좀 알려줘
  2. 환율을 계산할 때 - 15.5 유로가 얼마야? 
  3. 알람을 설정할 때 - 오전 6시에 알람맞춰줘  

*  모두가 알고 있는 시리의 상세소개 및 작동법은 넘어가고 몰랐던 사실 몇 가지를 소개하자면
  1. 2011년 서비스가 시작되어 벌써 9년 차다
  2. 2019년 8월 그레이딩 (평가프로그램) 과정에서 외부인력이 녹음 파일을 들은 것에 대해 공식 사과했다. (직업 윤리와 관련)
  3. 시리의 음성인식 엔진 제공 회사인 뉘앙스 커뮤니케이션은 아이폰과 안드로이드폰에 모두 앱을 출시하고 있다.  

* 그럼 시리는 어떻게 작동할까? 크게 세 가지가 있는데 우리가 "__아침 7시에 깨워줘__"라고 했을 때 어떤 일이 일어나는지 확인해보자.
  1. 음성 인식
    - 당신의 오디오를 해당 텍스트 형태인 "아침 7시에 깨워줘"로 변환한다.
    - 애플사의 엔지니어들은 최고의 음성 인식 모델을 만들기 위해 대규모의 전사 데이터 세트로 머신러닝 모델을 훈련시킨다. 
  2. 자연어 처리
    - 변환된 텍스트는 애플 서버로 보내져 추가 처리를 하게 된다. 이어 애플 서버는 이 텍스트에 NLP(Natural Language Processing) 알고리즘을 실행해 사용자가 말하고자 하는 의도를 파악한다.
    - "아침 7시에 깨워줘"라고 말할 때 사용자가 알람 설정에 대해 묻는 것이지 통화에 대해 묻는 것이 아니라는 점을 구별할 수 있다.
  3. 개체 추출
    - 문맥 이해의 기술이다. 인간에게 말하는 것처럼 시리에게 말할 수 있다.
    - You: 시리야, 알람 맞춰.  
시리: 몇 시에 알람 맞춰줄까?  
You: 오전 7시.

##3. dish-covery (구글맵)

* 해외에서 지내다 보면 수시로 켜는게 구글맵이다. 구글맵의 AI 및 머신러닝 사용을 조사하던 중 내가 __미처 모르면서 자주 이용했던 서비스__가 있어 놀랐다.

* 식당을 갈 때면 구글맵에 들어가 식당 클릭 -> 메뉴를 훑어보며 "여긴 뭐가 맛있지?" 고민 -> 리뷰 속 메뉴들 일일히 비교  

* 전에는 사장님들이 직접 찍어 올리는 줄 알았던 메뉴섹션은 사실 인공지능이 만들고 있었다. 이 서비스는
  1. 사람들의 리뷰에 들어가 있는 음식 사진 비교 및 대조하여 이름 __라벨링__
  2. 레스토랑의 __인기 있는 아이템__ 부터 내림차순 정렬
  3. 아이템을 누르면 그 __리뷰들을 종합__하여 제공

##4. 서울과기대 자율주행 셔틀

* 학교에서 신기한 서비스를 제공한다.  
코로나 온라인 수강 + 교환학생 으로 이용은 못하고 있지만  
다음학기부터 애용하게 될 것 같다 (과기대는 캠퍼스가 꽤 큰편)

* 과기대 셔틀버스의 자율 주행 단계는 4단계이다. 
![자율주행 단계 출처: 현대자동차그룹](http://res.heraldm.com/content/image/2019/05/17/20190517000630_0.jpg)  

* 어떻게 작동할까?
  1. 항상 새로운 곳으로 움직이므로 수많은 데이터들을 모을 수 밖에 없다.
  2. 기본적으로 '인식' - '판단' - '제어'로 작동된다.
  3. 딥러닝이 적용된다.  
    1. 시각 인식 지능의 적용
      - 전방 충돌 방지, 차선 이탈 방지, 차간 거리 조절 등 지능형 주행 및 자율 주행과 관련한 모든 기능은 주변 상황을 인식하는 것에서부터 시작
      - 인간은 시각 정보에 의해 운전하지만, 자율 주행 자동차는 카메라, 레이더, 라이다, 초음파, 적외선 등 다양한 정보를 복합적으로 활용  
    2. 학습 지능의 적용
      - 강화 학습(Reinforcement Learning), 관계형 추론(Relational Networks), 지능 이식(Transferring Intelligence)과 같은 기술을 자율 주행에 접목시키기 위한 연구가 진행 중
